### Exploratory Data Analysis(EDA)

```python
df.replace("?", pd.NA, inplace=True)
```

```python
df.dropna(subset=['Education', 'Annual Income'], inplace=True)
```

```python
df['Profession'].value_counts()
```

```python
df.groupby("Gender")['Anual Income'].mean()
```

```python
print(df[(df['Personal_loan'] == 'Yes') & (df['Marital_status'] == 'No')])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot: number of customers by profession
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Profession', order=df['Profession'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Customers by Profession")
plt.show()

# Pie chart: Distribution of Car Type
plt.figure(figsize=(6,6))
df['Car Type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Distribution of Car Type")
plt.ylabel('')
plt.show()

# Box plot: Annual income by education
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='Education', y='Annual Income')
plt.title("Income Distribution by Education Level")
plt.show()
```

```python
#Correlation &insight

# For numeric columns

correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
